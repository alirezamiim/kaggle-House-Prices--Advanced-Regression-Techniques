```python
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
train_data = pd.read_csv('data/train.csv',index_col='Id')
```

```python
def Change_to_int(column_name = '' , df = pd.DataFrame):
    helper_list = train_data[column_name].unique().tolist()
    for item in range(len(helper_list)):
        df[column_name][df[column_name]==helper_list[item]] = item
    return df


def Preparing_Data(train_data = pd.DataFrame,train_df = True):
#### MSSubClass    
    train_data['MSSubClass'] /= 10 
    train_data['MSSubClass'] -= train_data['MSSubClass'].mean()

#### MSZoning
    column_name = 'MSZoning'
    helper_list = train_data[column_name].unique().tolist()
    for item in range(len(helper_list)):
        train_data[column_name][train_data[column_name]==helper_list[item]] = item
   
    
    


#### GarageType 
    column_name = 'GarageType'
    train_data['GarageType'] = train_data['GarageType'].fillna(value='No Garage')
    helper_list = train_data[column_name].unique().tolist()
    for item in range(len(helper_list)):
        train_data[column_name][train_data[column_name]==helper_list[item]] = item

#### GarageYrBlt
    train_data['GarageYrBlt'] = train_data['YearBuilt'].max() - train_data['GarageYrBlt'] 
    train_data['GarageYrBlt'] = train_data['GarageYrBlt'].fillna(value=-1) 
    train_data['GarageYrBlt'] = train_data['GarageYrBlt']*2/train_data['GarageYrBlt'].max() - 1



#### GarageFinish
    column_name = 'GarageFinish'
    train_data[column_name] = train_data[column_name].fillna('No Garage')
    helper_list = train_data[column_name].unique().tolist()
    for item in range(len(helper_list)):
        train_data[column_name][train_data[column_name]==helper_list[item]] = item
    


#### GarageQual
    column_name = 'GarageQual'
    train_data[column_name] = train_data[column_name].fillna('No Garage')
    helper_list = train_data[column_name].unique().tolist()
    for item in range(len(helper_list)):
        train_data[column_name][train_data[column_name]==helper_list[item]] = item


#### GarageCond
    column_name = 'GarageCond'
    train_data[column_name] = train_data[column_name].fillna('No Garage')
    helper_list = train_data[column_name].unique().tolist()
    for item in range(len(helper_list)):
        train_data[column_name][train_data[column_name]==helper_list[item]] = item
    

#### GarageFinish
    column_name = 'GarageFinish'
    train_data[column_name] = train_data[column_name].fillna('No Garage')
    helper_list = train_data[column_name].unique().tolist()
    for item in range(len(helper_list)):
        train_data[column_name][train_data[column_name]==helper_list[item]] = item

#### GarageCars
    column_name = 'GarageCars'
    train_data[column_name] = train_data[column_name].fillna(value=0)

### GarageArea
    column_name = 'GarageArea'
    train_data[column_name] = train_data[column_name].fillna(value=0)
    train_data[column_name] = np.log2(train_data[column_name]+1)

#### WoodDeckSF
    column_name = 'WoodDeckSF'
    train_data[column_name] = np.log(train_data[column_name]+1)

#### OpenPorchSF
    column_name = 'OpenPorchSF'
    train_data[column_name] = np.log2(train_data[column_name]+1)




#### PavedDrive
    column_name = 'PavedDrive'
    helper_list = train_data[column_name].unique().tolist()
    for item in range(len(helper_list)):
        train_data[column_name][train_data[column_name]==helper_list[item]] = item





#### LotFrontage
    train_data['LotFrontage'] = train_data['LotFrontage'].fillna(value=np.random.randint(low=59,high=80))
    train_data['LotFrontage'] = np.log(train_data['LotFrontage'] )
    train_data['LotFrontage'] -= train_data['LotFrontage'].mean()

#### BsmtQual
    train_data['BsmtQual'] = train_data['BsmtQual'].fillna(value='No Basement')

#### BsmtCond
    train_data['BsmtCond'] = train_data['BsmtCond'].fillna(value='No Basement')

#### BsmtExposure
    train_data['BsmtExposure'] = train_data['BsmtExposure'].fillna(value='No Basement')


#### BsmtFinType1
    train_data['BsmtFinType1'] = train_data['BsmtFinType1'].fillna(value='No Basement')


#### BsmtFinType2
    train_data['BsmtFinType2'] = train_data['BsmtFinType2'].fillna(value='No Basement')

#### BsmtFinSF1
    train_data['BsmtFinSF1'] = np.log10(train_data['BsmtFinSF1']+1)

#### BsmtFinSF2
    train_data['BsmtFinSF2'] = np.log10(train_data['BsmtFinSF2']+1)

#### BsmtUnfSF
    train_data['BsmtUnfSF'] = np.log10(train_data['BsmtUnfSF']+1)

#### TotalBsmtSF
    train_data['TotalBsmtSF'] = np.log10(train_data['TotalBsmtSF']+1)

#### 1stFlrSF
    train_data['1stFlrSF'] = np.log10(train_data['1stFlrSF']+1)

#### 2ndFlrSF
    train_data['2ndFlrSF'] = np.log10(train_data['2ndFlrSF']+1)

#### GrLivArea
    train_data['GrLivArea'] = np.log10(train_data['GrLivArea']+1)

#### TotRmsAbvGrd
    train_data['TotRmsAbvGrd'] /= train_data['TotRmsAbvGrd'].max()









#### LotArea
    train_data['LotArea'] = np.log(train_data['LotArea'])
    train_data['LotArea'] = np.log10(train_data['LotArea'])
    train_data['LotArea'] -=3


#### Street
    train_data['Street'][train_data['Street']=='Grvl'] = 0
    train_data['Street'][train_data['Street']=='Pave'] = 1


#### Alley
    train_data['Alley'] = train_data['Alley'].fillna(value=0)
    train_data['Alley'][train_data['Alley']=='Grvl'] = 1
    train_data['Alley'][train_data['Alley']=='Pave'] = -1


#### MasVnrType
    train_data['MasVnrType'] = train_data['MasVnrType'].fillna(value = 'None')

    helper_list = train_data['MasVnrType'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['MasVnrType'][train_data['MasVnrType']==helper_list[item]] = item

#### MasVnrArea
    train_data['MasVnrArea'] = train_data['MasVnrArea'].fillna(value = 0)
    train_data['MasVnrArea'] +=1
    train_data['MasVnrArea'] = np.log(train_data['MasVnrArea'])



#### FireplaceQu
    train_data['FireplaceQu'] = train_data['FireplaceQu'].fillna(value = 0)
    column_name = 'FireplaceQu'
    helper_list = train_data[column_name].unique().tolist()
    for item in range(len(helper_list)):
        train_data[column_name][train_data[column_name]==helper_list[item]] = item
    
#### LotShape
    train_data['LotShape'][train_data['LotShape']=='Reg'] = -1
    train_data['LotShape'][train_data['LotShape']=='IR1'] = 0
    train_data['LotShape'][train_data['LotShape']=='IR2'] = 1
    train_data['LotShape'][train_data['LotShape']=='IR3'] = 2

#### LandContour    
    helper_list = train_data['LandContour'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['LandContour'][train_data['LandContour']==helper_list[item]] = item


#### Utilities    
    helper_list = train_data['Utilities'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['Utilities'][train_data['Utilities']==helper_list[item]] = item
    train_data['Utilities'] = train_data['Utilities'].fillna(value=np.random.randint(low=0,high=5))
    

#### LotConfig
    helper_list = train_data['LotConfig'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['LotConfig'][train_data['LotConfig']==helper_list[item]] = item
    

#### LandSlope
    train_data['LandSlope'][train_data['LandSlope']=='Gtl'] = -1
    train_data['LandSlope'][train_data['LandSlope']=='Mod'] = 0
    train_data['LandSlope'][train_data['LandSlope']=='Sev'] = 1

#### Electrical
    train_data['Electrical'] = train_data['Electrical'].fillna(value = train_data['Electrical'].describe().top)
    helper_list = train_data['Electrical'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['Electrical'][train_data['Electrical']==helper_list[item]] = item    


#### Neighborhood
    helper_list = train_data['Neighborhood'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['Neighborhood'][train_data['Neighborhood']==helper_list[item]] = item
    
    train_data['Neighborhood'] = train_data['Neighborhood']/12 - 1



#### Condition1
    helper_list = train_data['Condition1'].unique().tolist()
    for condition in range(len(helper_list)):
        train_data['Condition1'][train_data['Condition1']==helper_list[condition]] = condition

#### Condition2
    helper_list = train_data['Condition2'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['Condition2'][train_data['Condition2']==helper_list[item]] = item


#### BldgType
    helper_list = train_data['BldgType'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['BldgType'][train_data['BldgType']==helper_list[item]] = item



#### HouseStyle

    helper_list = train_data['HouseStyle'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['HouseStyle'][train_data['HouseStyle']==helper_list[item]] = item

#### OverallQual
    train_data['OverallQual'] -= train_data['OverallQual'].mean()



#### OverallCond
    train_data['OverallCond'] -= train_data['OverallCond'].mean()
    
#### YearBuilt
    last_year = train_data['YearBuilt'].max()
    train_data['YearBuilt'] = last_year - train_data['YearBuilt']
    train_data['YearBuilt'] = train_data['YearBuilt']*2/train_data['YearBuilt'].max() - 1


#### YearRemodAdd
    train_data['YearRemodAdd'] = last_year - train_data['YearRemodAdd']
    train_data['YearRemodAdd'][train_data['YearRemodAdd']>train_data['YearBuilt']] = 1
    train_data['YearRemodAdd'][train_data['YearRemodAdd']<=train_data['YearBuilt']] = 0


#### RoofStyle
    roof_style = train_data['RoofStyle'].unique().tolist()
    for style in range(len(roof_style)):
        train_data['RoofStyle'][train_data['RoofStyle']==roof_style[style]] = style


#### RoofMatl
    RoofMatl = train_data['RoofMatl'].unique().tolist()
    for material in range(len(RoofMatl)):
        train_data['RoofMatl'][train_data['RoofMatl']==RoofMatl[material]] = material


#### Exterior1st
    helper_list = train_data['Exterior1st'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['Exterior1st'][train_data['Exterior1st']==helper_list[item]] = item


    
#### Exterior2nd
    helper_list = train_data['Exterior2nd'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['Exterior2nd'][train_data['Exterior2nd']==helper_list[item]] = item

#### ExterQual
    helper_list = train_data['ExterQual'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['ExterQual'][train_data['ExterQual']==helper_list[item]] = item

#### ExterCond
    helper_list = train_data['ExterCond'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['ExterCond'][train_data['ExterCond']==helper_list[item]] = item


#### Foundation
    helper_list = train_data['Foundation'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['Foundation'][train_data['Foundation']==helper_list[item]] = item

#### BsmtQual
    helper_list = train_data['BsmtQual'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['BsmtQual'][train_data['BsmtQual']==helper_list[item]] = item

#### BsmtCond
    helper_list = train_data['BsmtCond'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['BsmtCond'][train_data['BsmtCond']==helper_list[item]] = item

#### BsmtExposure
    helper_list = train_data['BsmtExposure'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['BsmtExposure'][train_data['BsmtExposure']==helper_list[item]] = item

#### BsmtFinType1
    helper_list = train_data['BsmtFinType1'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['BsmtFinType1'][train_data['BsmtFinType1']==helper_list[item]] = item

#### BsmtFinType2
    helper_list = train_data['BsmtFinType2'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['BsmtFinType2'][train_data['BsmtFinType2']==helper_list[item]] = item 

#### Heating
    helper_list = train_data['Heating'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['Heating'][train_data['Heating']==helper_list[item]] = item

#### HeatingQC
    helper_list = train_data['HeatingQC'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['HeatingQC'][train_data['HeatingQC']==helper_list[item]] = item


#### CentralAir
    helper_list = train_data['CentralAir'].unique().tolist()
    for item in range(len(helper_list)):
        train_data['CentralAir'][train_data['CentralAir']==helper_list[item]] = item





#### KitchenQual
    column_name = 'KitchenQual'
    helper_list = train_data[column_name].unique().tolist()
    for item in range(len(helper_list)):
        train_data[column_name][train_data[column_name]==helper_list[item]] = item
    train_data[column_name] = train_data[column_name].fillna(value = np.random.randint(low=0,high=5))

#### BsmtHalfBath
    train_data['BsmtHalfBath'] = train_data['BsmtHalfBath'].fillna(value = 0)

#### BsmtFullBath
    train_data['BsmtFullBath'] = train_data['BsmtFullBath'].fillna(value = 0)


#### Functional
    column_name = 'Functional'
    helper_list = train_data[column_name].unique().tolist()
    for item in range(len(helper_list)):
        train_data[column_name][train_data[column_name]==helper_list[item]] = item
    train_data[column_name] = train_data[column_name].fillna(value = np.random.randint(low=0,high=9))



#### PoolArea
    column_name = 'PoolArea'
    train_data[column_name] = train_data[column_name].fillna(value=0)


#### PoolQC
    column_name = 'PoolQC'
    train_data[column_name] = train_data[column_name].fillna(value='No Pool')
    helper_list = train_data[column_name].unique().tolist()
    for item in range(len(helper_list)):
        train_data[column_name][train_data[column_name]==helper_list[item]] = item

#### Fence
    column_name = 'Fence'
    train_data[column_name] = train_data[column_name].fillna(value='No Fence')
    helper_list = train_data[column_name].unique().tolist()
    for item in range(len(helper_list)):
        train_data[column_name][train_data[column_name]==helper_list[item]] = item
    


#### MiscFeature
    column_name = 'MiscFeature'
    train_data[column_name] = train_data[column_name].fillna(value='No Feature')
    helper_list = train_data[column_name].unique().tolist()
    for item in range(len(helper_list)):
        train_data[column_name][train_data[column_name]==helper_list[item]] = item


#### YrSold

    column_name = 'YrSold'
    train_data[column_name] = last_year - train_data[column_name]



#### SaleCondition
    column_name = 'SaleCondition'
    helper_list = train_data[column_name].unique().tolist()
    for item in range(len(helper_list)):
        train_data[column_name][train_data[column_name]==helper_list[item]] = item
    
#### SaleType
    column_name = 'SaleType'
    helper_list = train_data[column_name].unique().tolist()
    for item in range(len(helper_list)):
        train_data[column_name][train_data[column_name]==helper_list[item]] = item
    train_data[column_name] = train_data[column_name].fillna(value = np.random.randint(low=0,high=11))


#### EnclosedPorch
    column_name = 'EnclosedPorch'
    train_data[column_name] /= train_data[column_name].max()



#### SalePrice
    if train_df == True:
        train_data['SalePrice'] /= 1000
        train_data['SalePrice'] = np.log2(train_data['SalePrice'])
        train_data['SalePrice'] -= 7
    


    return train_data
    




```

```python
train_data = pd.read_csv('data/train.csv',index_col='Id')
# print(train_data['SaleCondition'].unique())

column_list = train_data.columns.tolist()


train_data = Preparing_Data(train_data=train_data)
# train_data['MSZoning'][train_data['MSZoning']=='C (all)']
# print(train_data['MSZoning'].unique())
# for col in column_list:
#     print(col,train_data[col].dtypes,train_data[col].unique())

target = train_data['SalePrice']

features = train_data.drop(columns='SalePrice')

```
