```python
def MinMaxing(df=pd.DataFrame):
    column_list = df.columns.tolist()
    for col in column_list:
        min = df[col].min()
        max = df[col].max()
        df[col] = (df[col]-min)/(max-min)

    

    return df


def Preparing_Target(df = pd.DataFrame):
    df = np.log10(df/1000)
    df = df/df.max()
    return df
```
